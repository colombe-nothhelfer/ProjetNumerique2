### Question 6 : Euler explicite

````python
def solve_euler_explicit(f, x0, dt, t0, tf) :
    temps = [t0]
    positions = [x0]
    while t0 + dt < tf: 
        x0 = x0 + dt*f(t0, x0)
        positions.append(x0)
        t0 += dt
        temps.append(t0)
    return(np.array(temps), np.array(positions))
````

On teste notre fonction ```solve_euler_explicit``` sur l'équation différentielle $\dot x = -tx$...

````python
def f(t, x):
    return(-t*x)
````

...dont les solutions sont connues, ce sont les $x : t \mapsto C\mathrm{e}^{-t^2/2}$ pour $C \in \mathbb{R}$

````python
def g(t):
    return np.exp(-t**2/2)
````

````python
T_1, X_1 = solve_euler_explicit(f, 1, 0.25, 0, 3)

T_2, X_2 = solve_euler_explicit(f, 1, 10**(-1), 0, 3)

T_3, X_3 = solve_euler_explicit(f, 1, 10**(-2), 0, 3)

plt.plot(T_3, g(T_3), linestyle = '-', color='k')

plt.plot(T_1, X_1, linestyle='-.', label='$dt=0,25$')
plt.plot(T_2, X_2, linestyle='-.', label='$dt=0,1$')
plt.plot(T_3, X_3, linestyle='-.', label='$dt=0,01$')

plt.legend(loc=0)
plt.xlabel('$t$')
plt.show()   
````

Un schéma est dit convergent à l'ordre $p \in \mathbb{N^*}$ si $\underset{j}{max} \|x_j - x(t_j)\| \underset{dt \to 0}{\sim} dt^p$. Dans ce cas, on aura alors $\ln( \underset{j}{max} \|x_j - x(t_j)\| ) \underset{dt \ll 1}{\simeq} p \ln(dt)$. En traçant le graphe de $\ln( \underset{j}{max} \|x_j - x(t_j)\| )$ en fonction de $\ln(dt)$, on obtient, pour des valeurs faibles de $dt$, une droite de pente $p$.

````python
dt_liste = []
maximum = []

for i in range(1, 10):
    dt = 2**(-i)
    T, X = solve_euler_explicit(f, 1, dt, 0, 3)
    dt_liste.append(dt)
    m = max(abs(x - g(t)) for x, t in zip(X, T))
    maximum.append(m)
    
plt.plot(np.log(np.array(dt_liste)),
         np.log(np.array(maximum)),
         label='pente = ordre de convergence',
         linestyle='--',
         marker='o')
plt.legend(loc=0)
plt.xlabel('$\ln(dt)$')
plt.ylabel('$\ln( {max} \|x_j - x(t_j)\| )$')
plt.axis("equal")
plt.show()
````

### Question 7 : Lotka-Voltera explicite

````python
alpha = 2/3
beta = 4/3 
gamma = 1
delta = 1

def Lotka(t, x):
    x1, x2 = x
    return(np.array([x1*(alpha - beta*x2), -x2*(gamma - delta*x1)]))

T, X = solve_euler_explicit(Lotka, np.array([1, 1]), 10**(-2), 0, 100)
X1 = X[:, 0]
X2 = X[:, 1]
plt.plot(T, X1)
plt.plot(T, X2)
plt.show()
````

## Euler implicite

````python
eps = 10**(-3)

def solve_euler_implicit(f, x0, dt, t0, tf, itermax=100):
    
    temps = [t0]
    positions = [x0]
    
    while t0 + dt < tf:
        
        #calcul de x_{j+1}
        def F(x):
            return x0 + dt*f(t0, x)
        u, v = x0, F(x0)
        i = 0
        while (i < itermax) and (np.linalg.norm(u - v) > eps):
            u, v = v, F(v)
            i += 1
        x0 = u
        
        #comme euler normal
        positions.append(x0)
        t0 += dt
        temps.append(t0)
        
    return np.array(temps), np.array(positions)
````

````python
def g(t):
    return np.exp(-t**2/2)
T, X = solve_euler_implicit(f, 1, 10**(-1), 0, 3)
plt.plot(T, X)
plt.plot(T, g(T), linestyle = ':')
plt.show()

T, X = solve_euler_implicit(f, 1, 10**(-2), 0, 3)
plt.plot(T, X)
plt.plot(T, g(T), linestyle = ':')
plt.show()   
````
