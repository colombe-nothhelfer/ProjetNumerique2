```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def solve_euler_explicit(f, x0, dt, t0, tf) :
    temps = [t0]
    positions = [x0]
    while t0 + dt < tf: 
        x0 = x0 + dt*f(t0, x0)
        positions.append(x0)
        t0 += dt
        temps.append(t0)
    return(np.array(temps), np.array(positions))
```

```python
def f(t, x):
    return(-t*x)

solve_euler_explicit(f, 1, 10**(-2), 0, 3)
```

```python
def g(t):
    return np.exp(-t**2/2)
T, X = solve_euler_explicit(f, 1, 10**(-1), 0, 3)
plt.plot(T, X)
plt.plot(T, g(T), linestyle = ':')
plt.show()

T, X = solve_euler_explicit(f, 1, 10**(-2), 0, 3)
plt.plot(T, X)
plt.plot(T, g(T), linestyle = ':')
plt.show()

    
```

```python
dt_liste = []
maximum = []
for i in range(1, 10):
    dt = 2**(-i)
    T, X = solve_euler_explicit(f, 1, dt, 0, 3)
    dt_liste.append(dt)
    m = max(abs(x - g(t)) for x, t in zip(X, T))
    maximum.append(m)
plt.plot(np.log(np.array(dt_liste)), np.log(np.array(maximum)))
plt.axis("equal")
plt.show()
```

```python
alpha = 2/3
beta = 4/3 
gamma = 1
delta = 1
def Lotka(t, x):
    x1, x2 = x
    return(np.array([x1*(alpha - beta*x2), -x2*(gamma - delta*x1)]))

T, X = solve_euler_explicit(Lotka, np.array([1, 1]), 10**(-2), 0, 100)
X1 = X[:, 0]
X2 = X[:, 1]
plt.plot(T, X1)
plt.plot(T, X2)
plt.show()
```
