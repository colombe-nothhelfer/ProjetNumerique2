### Contexe

On considère le système différentiel suivant :
$$
\begin{aligned} (\mathcal{LV}) : \dot X = f(t,X) ~ \text{, où} ~ f : ~ & \mathbb{R} \times \mathbb{R}^2 & \longrightarrow ~ & \mathbb{R}^2 \\ & \left(t, \begin{pmatrix} x_1 \\ x_2 \end{pmatrix} \right) & \longmapsto ~ & \begin{pmatrix} x_1 (\alpha - \beta x_2) \\ - x_2 (\gamma - \delta x_1) \end{pmatrix} \end{aligned}
$$

avec $\alpha, \beta, \gamma, \delta$ des réels **strictement positifs** (On remarque que $\mathbb{R}$ et $\mathbb{R}^2$ sont bien des ouverts, respectivement d'eux-mêmes, et que $f$ est bien continue).

````python
alpha, beta, gamma, delta = 2/3, 4/3, 1, 1 #valeurs utilisées sur des articles de wikipédia
````

````python
import numpy as np
import matplotlib.pyplot as plt
````

### Question 1 : interprétation physique, points d'équilibre

##### la dynamique :

* $\dot x_1$ et $\dot x_2$ sont respectivement les variations de population de proies et de prédateurs.
* Concernant les proies :
    * Elles se reproduisent, donc leur population augmente proportionnellement à sa taille, d'où le terme $\alpha x_1$. Le facteur $\alpha$ représentant donc le nombre d'individus "produit" par chaque individu par unité de temps.
    * Elles sont mangées par les prédateurs. Le facteur $\beta$ représente la proportion de la population de proies mangée par chaque prédateur par unité de temps. Alors la population de proies diminue au cours du temps à cause de ce processus, d'où le terme $- \beta x_1 x_2$
* Concernant les prédateurs :
    * Sans aucune source de nourriture, ils meurent : la population diminue donc proportinnellement à sa taille, d'où le terme $- \gamma x_2$. Le facteur $\gamma$ représentant donc la proportion de la population qui décède chaque unité de temps.
    * Plus ils ont de nourriture à disposition, plus ils peuvent se reproduire : le terme $\delta x_2 x_1$ illustre donc le fait qu'à chaque unité de temps, un prédateur mange une proportion $\delta$ de la population de proies pour "produire" un prédateur de plus.

##### Les points d'équilibre

On cherche les points d'annulation de $f$. Si $(x_1, x_2) \in \mathbb{R}^2$ :

$$
f \begin{pmatrix} x_1 \\ x_2 \end{pmatrix} = \begin{pmatrix} 0 \\ 0 \end{pmatrix}
\Leftrightarrow
\left\{ \begin{aligned} x_1 (\alpha - \beta x_2) & = 0 \\ - x_2 (\gamma - \delta x_1) & = 0 \end{aligned} \right.
\Leftrightarrow
\left\{ \begin{aligned} x_1 & = 0 \\ x_2 & = 0 \end{aligned} \right.
\text{ou}
\left\{ \begin{aligned} x_1 & = \gamma / \delta \\ x_2 & = \alpha / \beta \end{aligned} \right.
$$

Il existe donc deux points d'équilibre : $(0,0)$ et $\bar x = (\frac{\gamma}{\delta}, \frac{\alpha}{\beta})$

$f$ est continûment différentiable et sa jacobienne vaut :
$$
\forall (x_1, x_2) \in \mathbb{R}, J_f(x_1, x_2) = \begin{pmatrix} \alpha - \beta x_2 & -\beta x_1 \\ \delta x_2 & \delta x_1 - \gamma \end{pmatrix}
$$

On peut alors étudier la stabilité des deux points d'équilibre :

* $J_f(0,0) = \begin{pmatrix} \alpha & 0 \\ 0 & -\gamma \end{pmatrix}$ donc $J_f(0,0)$ a une valeur propre strictement positive, donc **$(0,0)$ est instable**.

* $J_f(\bar x) = \begin{pmatrix} 0 & -\beta \frac{\gamma}{\delta} \\ \delta \frac{\alpha}{\beta} & 0 \end{pmatrix}$ et donc son polynome caractéristique est $X^2 + \alpha \gamma$. Les deux valeurs propres de $J_f(\bar x)$ sont des imaginaires purs, donc de partie réelle nulle : **on ne peut ainsi rien conclure, à ce stade, sur la stabilité de $\bar x$**

### Question 3 : Solutions "positives"

$f$ est continûment différentiable, et $\mathbb{R}^2$ est ouvert, donc par le théorème de Cauchy-Lipschitz, il y a unicité de la solution maximale passant par un point donné à un instant donné.

Soit $t_0 \in \mathbb{R}$, $x=(x_1, x_2)$ une solution maximale de $(\mathcal{LV})$ telle que $x(t_0) \in \mathbb{R}_{>0} \times \mathbb{R}_{>0}$. S'il existe $t_1 > t_0$ tel que $x(t_1) \notin \mathbb{R}_{>0} \times \mathbb{R}_{>0}$, alors on a $x_1(t_1) \leq 0$ ou $x_2(t_1) \leq 0$.
Dans le premier cas, par le théorème des valeurs intermédiaires, $x$ étant continue, il existe $t^* \in ]t_0, t_1[$ tel que $x_1(t^*) = 0$. Or $\tilde x : t \mapsto \begin{pmatrix} 0 \\ \mathrm{e}^{-\gamma t} \end{pmatrix}$ est une solution maximale de $(\mathcal{LV})$ qui n'est jamais dans $\mathbb{R}_{>0} \times \mathbb{R}_{>0}$ et qui pourtant vérifie bien $\tilde x_1(t^*) = 0$, ce qui contredit l'unicité. Dond il n'existe pas de tel $t_1$, et on en conclut que : **Toute solution initialisée dans $\mathbb{R}_{>0} \times \mathbb{R}_{>0}$ reste dans $\mathbb{R}_{>0} \times \mathbb{R}_{>0}$.**

### Question 4 : Une "énergie"

Soit $x=(x_1, x_2)$ une solution initialisée dans $\mathbb{R}_{>0} \times \mathbb{R}_{>0}$. On obtient, pour $t$ apprtenant au domaine de définition de $x$ :

$$
\begin{aligned}
\frac{\mathrm{d}}{\mathrm{d}t} ( \, H(x_1, x_2) \, ) & = \dot x_1 \left( \delta  - \frac{\gamma}{x_1} \right) + \dot x_2 \left( \beta - \frac{\alpha}{x_1} \right) \\ & = x_1 (\alpha - \beta x_2) \left( \delta  - \frac{\gamma}{x_1} \right) - x_2 (\gamma - \delta x_1) \left( \beta - \frac{\alpha}{x_2} \right) \\ & = (\alpha - \beta x_2)(\delta - \gamma x_1) - (\delta - \gamma x_1)(\alpha - \beta x_2) \\ & = 0
\end{aligned}
$$

Donc **$H$ est constante sur les trajectoires des solutions**, ou dit autrement, chaque solution de $(\mathcal{LV})$ est à valeur dans une courbe de niveau de $H$.
Or $H$ est continue donc, par les résultats du projet numérique n°1, ses courbes de niveau sont des compacts de $\mathbb{R}_{>0} \times \mathbb{R}_{>0}$, qui est un ouvert. Ainsi, si $\mathcal{C}$ est un compact de $\mathbb{R}_{>0} \times \mathbb{R}_{>0}$, il existe $\varepsilon > 0$ tel que $\forall x \in \mathcal{C}, d(x, \partial (\mathbb{R}_{>0} \times \mathbb{R}_{>0}) ) \ge \varepsilon$.

Considérons maintenant que $x$ est une solution maximale. $x$ est donc définie sur un intervalle ouvert $]t^-, t^+[ \subset \mathbb{R}$. Par ce qui précède, on ne peut avoir :
* ni $\underset{t \to t^+}{\mathrm{lim}} ~ d(~(t, x(t)), ~ \partial (\mathbb{R} \times (\mathbb{R}_{>0} \times \mathbb{R}_{>0}))~ ) = 0$,
* ni $\underset{t \to t^+}{\mathrm{lim}} ~ \|x(t)\| = + \infty$

Et ce, quelle que soit la valeur de $t^+$. 

Donc nécessairement, par le théorème de domaine maximal d'existence, $t^+ = + \infty$. Et de même, $t^- = - \infty$

Donc **toute solution maximale initialisée dans $\mathbb{R}_{>0} \times \mathbb{R}_{>0}$ est définie sur $\mathbb{R}$.**

### Question 5 : Courbes de niveau & stabilité

````python
def H(x1, x2):
    return delta * x1 - gamma * np.log(x1) + beta * x2 - alpha * np.log(x2)

H = np.vectorize(H)
````

````python
def display_contour(f, x, y, levels):
    X, Y = np.meshgrid(x, y)
    Z = f(X, Y)
    fig, ax = plt.subplots()
    contour_set = plt.contour(
        X, Y, Z, colors="grey", linestyles="dashed", 
        levels=levels 
    )
    ax.clabel(contour_set)
    plt.grid(True)
    plt.xlabel("$x_1$") 
    plt.ylabel("$x_2$")
    plt.gca().set_aspect("equal")
````

````python
display_contour(
    H, 
    x=np.linspace(0, 4.0, 100), 
    y=np.linspace(0, 2.0, 100), 
    levels=10 # 10 levels, automatically selected
)
plt.plot([gamma/delta], [alpha/beta], marker='*', color='r',
         label='$(\gamma / \delta, \alpha / \beta)$',
         linestyle='');
plt.legend(loc='upper right');
````

On remarque que $\bar x$ est un minimum de $H$. Ce qui justifie sa stabilité, à partir du moment qu'on a montré que $H$ était bien une fonction de Lyapunov.

### Question 6 : Schéma d'Euler explicite
##### La fonction

````python
def solve_euler_explicit(f, x0, dt, t0, tf) :
    
    temps = [t0]
    positions = [x0]
    
    while t0 + dt < tf: 
        
        x0 = x0 + dt*f(t0, x0)
        positions.append(x0)
        t0 += dt
        temps.append(t0)
    
    return np.array(temps), np.array(positions)
````

##### Test
On teste notre fonction ```solve_euler_explicit``` sur l'équation différentielle $\dot x = -tx$...

````python
def f(t, x):
    return(-t*x)
````

...dont les solutions sont connues, ce sont les $x : t \mapsto C\mathrm{e}^{-t^2/2}$ pour $C \in \mathbb{R}$ (en l'occurence, $C = 1$ dans le cas où on impose comme condition initiale $x(0) = 1$) :

````python
def g(t):
    return np.exp(-t**2/2)
````

````python
T_1_ee, X_1_ee = solve_euler_explicit(f, 1, 0.25, 0, 3)

T_2_ee, X_2_ee = solve_euler_explicit(f, 1, 10**(-1), 0, 3)

T_3_ee, X_3_ee = solve_euler_explicit(f, 1, 10**(-2), 0, 3)

plt.plot(T_3_ee, g(T_3_ee), linestyle = '-', color='k', label='$\mathrm{e}^{-t^2/2}$')

plt.plot(T_1_ee, X_1_ee, linestyle='--', label='$dt=0,25$')
plt.plot(T_2_ee, X_2_ee, linestyle='--', label='$dt=0,1$')
plt.plot(T_3_ee, X_3_ee, linestyle='--', label='$dt=0,01$')

plt.legend(loc=0)
plt.xlabel('$t$')
plt.ylabel('$x(t)$')
plt.show()   
````

###### Ordre de convergence 
Un schéma est dit convergent à l'ordre $p \in \mathbb{N^*}$ si $\underset{j}{max} \|x_j - x(t_j)\| \underset{dt \to 0}{\sim} dt^p$. Dans ce cas, on aura alors $\ln( \underset{j}{max} \|x_j - x(t_j)\| ) \underset{dt \ll 1}{\simeq} p \ln(dt)$. En traçant le graphe de $\ln( \underset{j}{max} \|x_j - x(t_j)\| )$ en fonction de $\ln(dt)$, on obtient, pour des valeurs faibles de $dt$, une droite de pente $p$.

````python
dt_liste = []
maximum = []

for i in range(1, 10):
    dt = 2**(-i)
    T, X = solve_euler_explicit(f, 1, dt, 0, 3)
    dt_liste.append(dt)
    m = max(abs(x - g(t)) for x, t in zip(X, T))
    maximum.append(m)
    
plt.plot(np.log(np.array(dt_liste)),
         np.log(np.array(maximum)),
         label='pente = ordre de convergence',
         linestyle='--',
         marker='o')
plt.legend(loc=0)
plt.xlabel('$\ln(dt)$')
plt.ylabel('$\ln( max ||x_j - x(t_j)|| )$')
plt.axis("equal")
plt.show()
````

On obtient bien une droite, de pente 1 : le schéma d'Euler explicite est donc convergent à l'ordre 1 (NB : le fait que ce graphe donne une droite illustre entre autres la convergence du schéma, qui se vérifie donc) 

### Question 7 : Lotka-Voltera explicite

````python
def Lotka(t, x):
    x1, x2 = x
    return(np.array([x1*(alpha - beta*x2), -x2*(gamma - delta*x1)]))

T_ee, X_ee = solve_euler_explicit(Lotka, np.array([1, 1]), 10**(-2), 0, 300)
X1_ee = X_ee[:, 0]
X2_ee = X_ee[:, 1]

plt.plot(T_ee[:2000], X1_ee[:2000], label='Proies')
plt.plot(T_ee[:2000], X2_ee[:2000], label='Prédateurs')
plt.xlabel('$t$')
plt.ylabel('Taille de la population [unité arbitraire]')
plt.legend(loc=0)
plt.show()

plt.plot(T_ee, X1_ee, label='Proies')
plt.plot(T_ee, X2_ee, label='Prédateurs')
plt.xlabel('$t$')
plt.ylabel('Taille de la population [unité arbitraire]')
plt.legend(loc=0)
plt.show()

plt.plot(T_ee, H(X1_ee, X2_ee), label='$H$')
plt.xlabel('$t$')
plt.axis(ymin=0)
plt.legend(loc=0)
plt.show()
````

Les solutions semblent bien présenter une certaine cyclicité, mais les maximums de population, des proies comme des prédateurs, augmentent pourtant au cours du temps. Cela ne semble pas cohérent avec la réalité, où l'on observe pas d'augmentation globale des populations.

De plus, lorsque l'on trace la valeur de $H$ le long de la trajectoire au cours du temps, on voit qu'elle augmente exponentiellement. Or la question 4 nous assure qu'elle doit rester constante...

### Question 8 : Schéma d'Euler implicite

##### La fonction

````python
eps = 10**(-7) 
#pas plus, car si eps est de l'ordre des dt utilisés dans la cellule ci-après
#pour tester la convergence, alors on a des petits problèmes (de convergence, justement)

def solve_euler_implicit(f, x0, dt, t0, tf, itermax=100):
    
    temps = [t0]
    positions = [x0]
    
    while t0 + dt < tf:
        
        #calcul de x_{j+1}
        def F(x):
            return x0 + dt*f(t0 + dt, x)
        u, v = x0, F(x0)
        i = 0
        while (i < itermax) and (np.linalg.norm(u - v) > eps):
            u, v = v, F(v)
            i += 1
        x0 = u
        
        #comme euler normal
        positions.append(x0)
        t0 += dt
        temps.append(t0)
        
    return np.array(temps), np.array(positions)
````

##### Test sur équation aux solutions exactes connues

````python
T_1_ei, X_1_ei = solve_euler_implicit(f, 1, 0.25, 0, 3)

T_2_ei, X_2_ei = solve_euler_implicit(f, 1, 10**(-1), 0, 3)

T_3_ei, X_3_ei = solve_euler_implicit(f, 1, 10**(-2), 0, 3)

plt.plot(T_3_ei, g(T_3_ei), linestyle = '-', color='k', label='$\mathrm{e}^{-t^2/2}$')

plt.plot(T_1_ei, X_1_ei, linestyle='--', label='$dt=0.25$')
plt.plot(T_2_ei, X_2_ei, linestyle='--', label='$dt=0.1$')
plt.plot(T_3_ei, X_3_ei, linestyle='--', label='$dt=0.01$')

plt.legend(loc=0)
plt.xlabel('$t$')
plt.ylabel('$x(t)$')
plt.show()  
````

````python
dt_liste = []
maximum = []

for i in range(2, 10):
    dt = 2**(-i)
    T, X = solve_euler_implicit(f, 1, dt, 0, 3)
    dt_liste.append(dt)
    m = max(abs(x - g(t)) for x, t in zip(X, T))
    maximum.append(m)
    
plt.plot(np.array(dt_liste),
         np.array(maximum),
         label='pente = ordre de convergence',
         linestyle='--',
         marker='o')
plt.legend(loc=0)
plt.xlabel('$\mathrm{d}t$')
plt.ylabel('$max~||x_j - x(t_j)||$')
plt.axis(xmin=0, ymin=0)
plt.show()

plt.plot(np.log(np.array(dt_liste)),
         np.log(np.array(maximum)),
         label='pente = ordre de convergence',
         linestyle='--',
         marker='o')
plt.legend(loc=0)
plt.xlabel('$\ln(dt)$')
plt.ylabel('$\ln( max ||x_j - x(t_j)|| )$')
plt.axis("equal")
plt.show()
````

Sur un système d'équations dont on connait des solutions analytiques, on remarque qu'on a affaire à un schéma convergent à un ordre 1 également.

##### Test sur les équations de Lotka- Volterra

````python
T, X = solve_euler_implicit(Lotka, np.array([1, 1]), 10**(-2), 0, 300)
X1 = X[:, 0]
X2 = X[:, 1]

plt.plot(T[:2000], X1[:2000], label='Proies')
plt.plot(T[:2000], X2[:2000], label='Prédateurs')
plt.xlabel('$t$')
plt.ylabel('Taille de la population [unité arbitraire]')
plt.legend(loc=0)
plt.show()

plt.plot(T, X1, label='Proies')
plt.plot(T, X2, label='Prédateurs')
plt.xlabel('$t$')
plt.ylabel('Taille de la population [unité arbitraire]')
plt.legend(loc=0)
plt.show()

plt.plot(T, H(X1, X2), label='$H$')
plt.xlabel('$t$')
plt.axis(ymin=0)
plt.legend(loc='lower left')
plt.show()
````

Cette fois-ci, le comportement des résultats change : H diminue un peu sur les trajectoires, et les populations de proies et prédateurs ne divergent plus, mais semblent au contraire converger (vers $\bar x$, qui plus est).

### Question 9 : un nouveau système pas si différent

Soit $u : \mathbb{R}^2 \to \mathbb{R}^2$ continûment différentiable et $H_0$ un réel. Considérons alors le système suivant :

$$
(\tilde{\mathcal{LV}}) : \dot X = \tilde f(X) ~ \text{, où} ~ \tilde f : \begin{pmatrix} x_1 \\ x_2 \end{pmatrix} \in \mathbb{R}^2 \mapsto  \begin{pmatrix} x_1 (\alpha - \beta x_2) - u_1(x_1, x_2)(H(x_1, x_2) - H_0) \\ - x_2 (\gamma - \delta x_1) - u_2(x_1, x_2)(H(x_1, x_2) - H_0) \end{pmatrix}
$$

**Si l'on choisit $H_0 = H(x(0))$**, alors puisque $H$ est une fonction constante le long des trajectoires des solutions (cf. question 4), on a que $H(x_1, x_2) - H_0 = 0$. Ce qui donne $(\tilde{\mathcal{LV}}) \Leftrightarrow (\mathcal{LV})$, donc **les solutions de $(\tilde{\mathcal{LV}})$ sont exactement celles de $(\mathcal{LV})$**, quel que soit le choix de $u$ continûment différentiable.
