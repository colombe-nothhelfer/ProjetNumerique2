### Contexe

On considère le système différentiel suivant :
$$
(\mathcal{E}) : \dot X = f(X) \text{, où} f : \begin{pmatrix} x_1 \\ x_2 \end{pmatrix} \in \mathbb{R}^2 \mapsto  \begin{pmatrix} x_1 (\alpha - \beta x_2) \\ - x_2 (\gamma - \delta x_1) \end{pmatrix}
$$

avec $\alpha, \beta, \gamma, \delta$ des réels **strictement positifs**.

````python
alpha, beta, gamma, delta = 2/3, 4/3, 1, 1
````

````python
import numpy as np
import matplotlib.pyplot as plt
````

### Question 1 : interprétation physique, points d'équilibre

##### la dynamique :

* $\dot x_1$ et $\dot x_2$ sont respectivement les variations de population de proies et de prédateurs.
* Concernant les proies :
    * Elles se reproduisent, donc leur population augmente proportionnellement à sa taille, d'où le terme $\alpha x_1$. Le facteur $\alpha$ représentant donc le nombre d'individus "produit" par chaque individu par unité de temps.
    * Elles sont mangées par les prédateurs. Le facteur $\beta$ représente la proportion de la population de proies mangée par chaque prédateur par unité de temps. Alors la population de proies diminue au cours du temps à cause de ce processus, d'où le terme $- \beta x_1 x_2$
* Concernant les prédateurs :
    * Sans aucune source de nourriture, ils meurent : la population diminue donc proportinnellement à sa taille, d'où le terme $- \gamma x_2$. Le facteur $\gamma$ représentant donc la proportion de la population qui décède chaque unité de temps.
    * Plus ils ont de nourriture à disposition, plus ils peuvent se reproduire : le terme $\delta x_2 x_1$ illustre donc le fait qu'à chaque unité de temps, un prédateur mange une proportion $\delta$ de la population de proies pour "produire" un prédateur de plus.

##### Les points d'équilibre

On cherche les points d'annulation de $f$. Si $(x_1, x_2) \in \mathbb{R}^2$ :

$$
f \begin{pmatrix} x_1 \\ x_2 \end{pmatrix} = \begin{pmatrix} 0 \\ 0 \end{pmatrix}
\Leftrightarrow
\left\{ \begin{aligned} x_1 (\alpha - \beta x_2) & = 0 \\ - x_2 (\gamma - \delta x_1) & = 0 \end{aligned} \right.
\Leftrightarrow
\left\{ \begin{aligned} x_1 & = 0 \\ x_2 & = 0 \end{aligned} \right.
\text{ou}
\left\{ \begin{aligned} x_1 & = \gamma / \delta \\ x_2 & = \alpha / \beta \end{aligned} \right.
$$

Il existe donc deux points d'équilibre : $(0,0)$ et $\bar x = (\frac{\gamma}{\delta}, \frac{\alpha}{\beta})$

$f$ est continûment différentiable et sa jacobienne vaut :
$$
\forall (x_1, x_2) \in \mathbb{R}, J_f(x_1, x_2) = \begin{pmatrix} \alpha - \beta x_2 & -\beta x_1 \\ \delta x_2 & \delta x_1 - \gamma \end{pmatrix}
$$

On peut alors étudier la stabilité des deux points d'équilibre :

* $J_f(0,0) = \begin{pmatrix} \alpha & 0 \\ 0 & -\gamma \end{pmatrix}$ donc $J_f(0,0)$ a une valeur propre strictement positive, donc **$(0,0)$ est instable**.

* $J_f(\bar x) = \begin{pmatrix} 0 & -\beta \frac{\gamma}{\delta} \\ \delta \frac{\alpha}{\beta} & 0 \end{pmatrix}$ et donc son polynome caractéristique est $X^2 + \alpha \gamma$. Les deux valeurs propres de $J_f(\bar x)$ sont des imaginaires purs, donc de partie réelle nulle : **on ne peut ainsi rien conclure, à ce stade, sur la stabilité de $\bar x$**

### Question 3 : Solutions "positives"

$f$ est continûment différentiable, et $\mathbb{R}^2$ est ouvert, donc par le théorème de Cauchy-Lipschitz, il y a unicité de la solution maximale passant par un point donné à un instant donné.

Soit $t_0 \in \mathbb{R}$, $x=(x_1, x_2)$ une solution maximale de $(\mathcal{E})$ telle que $x(t_0) \in \mathbb{R}_{>0} \times \mathbb{R}_{>0}$. S'il existe $t_1 > t_0$ tel que $x(t_1) \notin \mathbb{R}_{>0} \times \mathbb{R}_{>0}$, alors on a $x_1(t_1) \leq 0$ ou $x_2(t_1) \leq 0$.
Dans le premier cas, par le théorème des valeurs intermédiaires, $x$ étant continue, il existe $t^* \in ]t_0, t_1[$ tel que $x_1(t^*) = 0$. Or $\tilde x : t \mapsto \begin{pmatrix} 0 \\ \mathrm{e}^{-\gamma t} \end{pmatrix}$ est une solution maximale de $(\mathcal{E})$ qui n'est jamais dans $\mathbb{R}_{>0} \times \mathbb{R}_{>0}$ et qui pourtant vérifie bien $\tilde x_1(t^*) = 0$, ce qui contredit l'unicité. Dond il n'existe pas de tel $t_1$, et on en conclut que : **Toute solution initialisée dans $\mathbb{R}_{>0} \times \mathbb{R}_{>0}$ reste dans $\mathbb{R}_{>0} \times \mathbb{R}_{>0}$.**

### Question 4 : Une "énergie"

Soit $x=(x_1, x_2)$ une solution initialisée dans $\mathbb{R}_{>0} \times \mathbb{R}_{>0}$. On obtient, pour $t$ apprtenant au domaine de définition de $x$ :

$$
\begin{aligned}
\frac{\mathrm{d}}{\mathrm{d}t} ( \, H(x_1, x_2) \, ) & = \dot x_1 \left( \delta  - \frac{\gamma}{x_1} \right) + \dot x_2 \left( \beta - \frac{\alpha}{x_1} \right) \\ & = x_1 (\alpha - \beta x_2) \left( \delta  - \frac{\gamma}{x_1} \right) - x_2 (\gamma - \delta x_1) \left( \beta - \frac{\alpha}{x_2} \right) \\ & = (\alpha - \beta x_2)(\delta - \gamma x_1) - (\delta - \gamma x_1)(\alpha - \beta x_2) \\ & = 0
\end{aligned}
$$

Donc H est constante sur les trajectoires des solutions.

**[EN DEDUIRE QUE TOUTE SOL MAX INITIALISEE DANS $\mathbb{R}_{>0} \times \mathbb{R}_{>0}$ EST DEFINIE SUR $\mathbb{R}$]**

### Question 5 : Courbes de niveau & stabilité

````python
def H(x1, x2):
    return delta * x1 - gamma * np.log(x1) + beta * x2 - alpha * np.log(x2)
````

````python
def display_contour(f, x, y, levels):
    X, Y = np.meshgrid(x, y)
    Z = f(X, Y)
    fig, ax = plt.subplots()
    contour_set = plt.contour(
        X, Y, Z, colors="grey", linestyles="dashed", 
        levels=levels 
    )
    ax.clabel(contour_set)
    plt.grid(True)
    plt.xlabel("$x_1$") 
    plt.ylabel("$x_2$")
    plt.gca().set_aspect("equal")
````

````python
display_contour(
    H, 
    x=np.linspace(0.1, 4.0, 100), 
    y=np.linspace(0.1, 2.0, 100), 
    levels=10 # 10 levels, automatically selected
)
plt.plot([gamma/delta], [alpha/beta], marker='*', color='r',
         label='$(\gamma / \delta, \alpha / \beta)$',
         linestyle='');
plt.legend(loc='upper right');
````

### Question 6 : Schéma d'Euler explicite

````python
def solve_euler_explicit(f, x0, dt, t0, tf) :
    
    temps = [t0]
    positions = [x0]
    
    while t0 + dt < tf: 
        
        x0 = x0 + dt*f(t0, x0)
        positions.append(x0)
        t0 += dt
        temps.append(t0)
    
    return np.array(temps), np.array(positions)
````

On teste notre fonction ```solve_euler_explicit``` sur l'équation différentielle $\dot x = -tx$...

````python
def f(t, x):
    return(-t*x)
````

...dont les solutions sont connues, ce sont les $x : t \mapsto C\mathrm{e}^{-t^2/2}$ pour $C \in \mathbb{R}$ :

````python
def g(t):
    return np.exp(-t**2/2)
````

````python
T_1, X_1 = solve_euler_explicit(f, 1, 0.25, 0, 3)

T_2, X_2 = solve_euler_explicit(f, 1, 10**(-1), 0, 3)

T_3, X_3 = solve_euler_explicit(f, 1, 10**(-2), 0, 3)

plt.plot(T_3, g(T_3), linestyle = '-', color='k', label='$\mathrm{e}^{-t^2/2}$')

plt.plot(T_1, X_1, linestyle='--', label='$dt=0,25$')
plt.plot(T_2, X_2, linestyle='--', label='$dt=0,1$')
plt.plot(T_3, X_3, linestyle='--', label='$dt=0,01$')

plt.legend(loc=0)
plt.xlabel('$t$')
plt.ylabel('$x(t)$')
plt.show()   
````

Un schéma est dit convergent à l'ordre $p \in \mathbb{N^*}$ si $\underset{j}{max} \|x_j - x(t_j)\| \underset{dt \to 0}{\sim} dt^p$. Dans ce cas, on aura alors $\ln( \underset{j}{max} \|x_j - x(t_j)\| ) \underset{dt \ll 1}{\simeq} p \ln(dt)$. En traçant le graphe de $\ln( \underset{j}{max} \|x_j - x(t_j)\| )$ en fonction de $\ln(dt)$, on obtient, pour des valeurs faibles de $dt$, une droite de pente $p$.

````python
dt_liste = []
maximum = []

for i in range(1, 10):
    dt = 2**(-i)
    T, X = solve_euler_explicit(f, 1, dt, 0, 3)
    dt_liste.append(dt)
    m = max(abs(x - g(t)) for x, t in zip(X, T))
    maximum.append(m)
    
plt.plot(np.log(np.array(dt_liste)),
         np.log(np.array(maximum)),
         label='pente = ordre de convergence',
         linestyle='--',
         marker='o')
plt.legend(loc=0)
plt.xlabel('$\ln(dt)$')
plt.ylabel('$\ln( {max} \|x_j - x(t_j)\| )$')
plt.axis("equal")
plt.show()
````

On obtient bien une droite, de pente 1 : le schéma d'Euler explicite est donc convergent à l'ordre 1.

### Question 8 : Schéma d'Euler implicite

````python
eps = 10**(-3)

def solve_euler_implicit(f, x0, dt, t0, tf, itermax=100):
    
    temps = [t0]
    positions = [x0]
    
    while t0 + dt < tf:
        
        #calcul de x_{j+1}
        def F(x):
            return x0 + dt*f(t0, x)
        u, v = x0, F(x0)
        i = 0
        while (i < itermax) and (np.linalg.norm(u - v) > eps):
            u, v = v, F(v)
            i += 1
        x0 = u
        
        #comme euler normal
        positions.append(x0)
        t0 += dt
        temps.append(t0)
        
    return np.array(temps), np.array(positions)
````

````python
def g(t):
    return np.exp(-t**2/2)
T, X = solve_euler_implicit(f, 1, 10**(-1), 0, 3)
plt.plot(T, X)
plt.plot(T, g(T), linestyle = ':')
plt.show()

T, X = solve_euler_implicit(f, 1, 10**(-2), 0, 3)
plt.plot(T, X)
plt.plot(T, g(T), linestyle = ':')
plt.show()   
````
